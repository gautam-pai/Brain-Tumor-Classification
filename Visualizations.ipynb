```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

### Import Data

```python
adam_accuracy=pd.read_csv("Adam_Accuaracy.csv")
adam_losses=pd.read_csv("Adam_Losses.csv")
sgd_accuracy=pd.read_csv("SGD_Accuaracy.csv")
sgd_losses=pd.read_csv("SGD_Losses.csv")
```

```python
adam_accuracy=adam_accuracy.rename(columns={'Training Accuaracy-1':'Training Accuracy-1','Testing Accuaracy-1':'Testing Accuracy-1',
                      'Training Accuaracy-32':'Training Accuracy-32','Testing Accuaracy-32':'Testing Accuracy-32',
                      'Training Accuaracy-64':'Training Accuracy-64','Testing Accuaracy-64':'Testing Accuracy-64'})
```

### Comparison Batch Size = 1 

#### Accuracy

```python
# Training

plt.figure(figsize=(10,5))
plt.suptitle("Training Datset Performance", fontsize=20, y=1.02)

plt.subplot(1,2,1)
plt.plot(adam_accuracy['Training Accuracy-1'],label="Adam Accuracy")
plt.plot(sgd_accuracy['Training Accuracy-1'],label="SGD  Accuracy")
plt.title("Model Accuracy")
plt.legend()


plt.subplot(1,2,2)
plt.plot(adam_losses['Training Loss-1'],label="Adam Loss")
plt.plot(sgd_losses['Training Loss-1'],label="SGD Loss")
plt.title("Gradient Descent Training Loss")
plt.legend()

```

```python
# Validation

plt.figure(figsize=(10,5))
plt.suptitle("Test Datset Performance", fontsize=20,y=1.02)

plt.subplot(1,2,1)
plt.plot(adam_accuracy['Testing Accuracy-1'],label="Adam Accuracy")
plt.plot(sgd_accuracy['Testing Accuracy-1'],label="SGD  Accuracy")
plt.title("Model Accuracy on Validation Set")
plt.legend(loc="upper right")


plt.subplot(1,2,2)
plt.plot(adam_losses['Testing Loss-1'],label="Adam Loss")
plt.plot(sgd_losses['Testing Loss-1'],label="SGD Loss")
plt.title("Gradient Descent Training Loss")
plt.legend()
```

## Batch Size = 32

```python
# Training Set


plt.figure(figsize=(15,5))
plt.suptitle("Training Datset Performance", fontsize=20, y=1.02)

plt.subplot(1,3,1)
plt.plot(adam_accuracy['Training Accuracy-32'],label="Adam Accuracy")
plt.plot(sgd_accuracy['Training Accuracy-32'],label="SGD  Accuracy")
plt.title("Model Accuracy")
plt.legend()


plt.subplot(1,3,2)
plt.plot(adam_losses['Training Loss-32'],label="Adam Loss")
plt.plot(sgd_losses['Training Loss-32'],label="SGD Loss")
plt.title("Gradient Descent Validation Loss")
plt.legend()

plt.subplot(1,3,3)
plt.plot(sgd_losses['Training Loss-32'],label="SGD Loss Zoomed In",color="orange")

plt.legend()
plt.show()
```

```python
# Validation Set

plt.figure(figsize=(15,5))
plt.suptitle("Test Datset Performance", fontsize=20,y=1.02)
plt.subplot(1,3,1)
plt.plot(adam_accuracy['Testing Accuracy-32'],label="Adam Accuracy")
plt.plot(sgd_accuracy['Testing Accuracy-32'],label="SGD  Accuracy")
plt.title("Model Accuracy")
plt.legend()


plt.subplot(1,3,2)
plt.plot(adam_losses['Testing Loss-32'],label="Adam Loss")
plt.plot(sgd_losses['Testing Loss-32'],label="SGD Loss")
plt.title("Gradient Descent Training Loss")
plt.legend()

plt.subplot(1,3,3)
plt.plot(sgd_losses['Testing Loss-32'],label="SGD Loss Zoomed In",color="orange")
plt.legend()
plt.show()
```
